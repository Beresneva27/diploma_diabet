```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting style
sns.set_style("whitegrid")

import datetime as dt


def plot_glucose_for_given_user_with_df(init_df, user_id: int | None, x_lower=None, x_upper=None,
                                        id_col='id', time_col='time', glucose_col='glucose', with_hypo_glic_lines=False):
    if user_id is not None:
        user_df = init_df[init_df[id_col] == user_id].sort_values(by=time_col)
    else:
        user_df = init_df
    
    plt.figure(figsize=(20, 12))

    if x_lower is None:
        x_lower = min(user_df[time_col])
    else:
        x_lower = dt.datetime.fromisoformat(x_lower)
    
    if x_upper is None:
        x_upper = max(user_df[time_col])
    else:
        x_upper = dt.datetime.fromisoformat(x_upper) 

    if with_hypo_glic_lines:
        plt.hlines(y=10, xmin=x_lower, xmax=x_upper, colors='red', lw=2, label='Hyperglycemia')
        plt.hlines(y=3.3, xmin=x_lower, xmax=x_upper, colors='aqua', lw=2, label='Hypoglycemia')
        plt.axhspan(4.6, 7.8, alpha=0.1, color='green', label='Euglycemia')
    
    plt.xlim(x_lower, x_upper)
    plt.xlabel('Время')
    plt.ylabel('Глюкоза, ммоль/л')
    plt.plot(user_df[time_col], user_df[glucose_col], label='Glucose level')
    plt.title(f'Glucose level for patient with {id_col} = {user_id}')
    plt.legend()
    # plt.savefig('last_image.png')
    plt.show()
```

```python
plot_glucose_for_given_user_with_df(res_data[res_data['Glucose'].notna()], user_id=9, time_col='Time', id_col='patient_id', glucose_col='Glucose', with_hypo_glic_lines=True)
```

```python
def plot_glucose_with_insuling_and_carbs_for_given_user_with_df(init_df, user_id: int | None, x_lower=None, x_upper=None, y_lower=None, y_upper=None, with_hypo_glic_lines=False,
                                                                id_col='id', time_col='time', glucose_col='glucose', carbs_col='carbs', insulin_col='rapid_insulin'):
    if user_id is not None:
        user_df = init_df[init_df[id_col] == user_id].sort_values(by=time_col)
    else:
        user_df = init_df
    
    plt.figure(figsize=(20, 12))

    if x_lower is None:
        x_lower = min(user_df[time_col])
    else:
        x_lower = dt.datetime.fromisoformat(x_lower)
    
    if x_upper is None:
        x_upper = max(user_df[time_col])
    else:
        x_upper = dt.datetime.fromisoformat(x_upper) 


    if with_hypo_glic_lines:
        plt.hlines(y=10, xmin=x_lower, xmax=x_upper, colors='red', lw=2, label='Hyperglycemia')
        plt.hlines(y=3.3, xmin=x_lower, xmax=x_upper, colors='aqua', lw=2, label='Hypoglycemia')
        plt.axhspan(4.6, 7.8, alpha=0.1, color='green', label='Euglycemia')

    
    plt.scatter(user_df[time_col], user_df[carbs_col] / 2, label='Carbohydrates', color='skyblue')
    plt.scatter(user_df[time_col], user_df[insulin_col], label='Rapid Insulin', color='green')

    if y_lower is not None and y_upper is not None:
        plt.ylim(y_lower, y_upper)
        
    plt.xlim(x_lower, x_upper)
    plt.xlabel('Время')
    # plt.ylabel('Глюкоза, ммоль/л')
    plt.plot(user_df[user_df[glucose_col].notna()][time_col], user_df[user_df[glucose_col].notna()][glucose_col], label='Glucose level')
    plt.title(f'Glucose level for patient with {id_col} = {user_id}')
    plt.legend()
    plt.savefig('last_image.png')
    plt.show()
```

```python
plot_glucose_with_insuling_and_carbs_for_given_user_with_df(res_data, user_id=9, x_lower='2024-11-24', x_upper='2024-11-25', with_hypo_glic_lines=True, y_lower=0, y_upper=20,
                                                            time_col='Time', id_col='patient_id', glucose_col='Glucose', carbs_col='Carbohydrates', insulin_col='Rapid Insulin')
```

```python
sample_rate = 15
history = 3
roll_window_width = history * 60 // sample_rate

given_id = 9
df_corr = res_data[res_data['patient_id'] == given_id]


df_corr[f'calories {history}h'] = df_corr['calories'].rolling(roll_window_width).sum()
df_corr[f'calories {history}h'].replace(to_replace=np.nan, method='bfill', inplace=True)
df_corr[f'distance {history}h'] = df_corr['distance'].rolling(roll_window_width).sum()
df_corr[f'distance {history}h'].replace(to_replace=np.nan, method='bfill', inplace=True)
df_corr[f'bpm {history}h'] = df_corr['bpm'].rolling(roll_window_width).mean()
df_corr[f'bpm {history}h'].replace(to_replace=np.nan, method='bfill', inplace=True)

corr_3h = df_corr[["Glucose", f"calories {history}h", f"distance {history}h", f"bpm {history}h"]].corr(method="pearson")


history = 1

df_corr = res_data[res_data['patient_id'] == given_id]
df_corr[f'calories {history}h'] = df_corr['calories'].rolling(roll_window_width).sum()
df_corr[f'calories {history}h'].replace(to_replace=np.nan, method='bfill', inplace=True)
df_corr[f'distance {history}h'] = df_corr['distance'].rolling(roll_window_width).sum()
df_corr[f'distance {history}h'].replace(to_replace=np.nan, method='bfill', inplace=True)
df_corr[f'bpm {history}h'] = df_corr['bpm'].rolling(roll_window_width).mean()
df_corr[f'bpm {history}h'].replace(to_replace=np.nan, method='bfill', inplace=True)

corr_1h = df_corr[["Glucose", f"calories {history}h", f"distance {history}h", f"bpm {history}h"]].corr(method="pearson")
```

```python
f, axes = plt.subplots(1, 2, figsize=(22,10))
label_size, tick_size, title_size = 16, 16, 19

sns.set(font_scale=2)
for idx, duration in enumerate([3, 1]):
    if duration == 1:
        mask = np.zeros_like(corr_1h)
        corr = corr_1h
    elif duration == 3:
        mask = np.zeros_like(corr_3h)
        corr = corr_3h
    mask[np.triu_indices_from(mask)] = True
    with sns.axes_style("white"):
        sns.heatmap(corr, mask=mask, square=True, annot=True, ax=axes[idx])
        axes[idx].set_title(f"{duration}h correlations", fontsize=title_size)
        axes[idx].tick_params(axis='both', which='major', labelsize=label_size)
f.tight_layout(pad=0.5)
# plt.savefig('last_image.png')
```

```python
import statsmodels.tsa.api as smt
import statsmodels.api as sm

plt.rcParams.update({'figure.max_open_warning': 0})

sns.set()

def tsplot(y, lags=None, figsize=(12, 7), style='bmh', given_id=None):
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):    
        fig = plt.figure(figsize=figsize)
        layout = (3, 2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0), colspan=2)
        pacf_ax = plt.subplot2grid(layout, (2, 0), colspan=2)

        y.plot(ax=ts_ax)
        if given_id is not None:
            ts_ax.set_title(f'Time Series Analysis Plots for patient_id = {given_id}')
        else:
            ts_ax.set_title('Time Series Analysis Plots')
        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.5)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.5)

        dickey_fuller_pvalue = sm.tsa.stattools.adfuller(y)[1]
        if given_id is not None:
            print(f"Dickey—Fuller test for patient_id = {given_id}:\tp = {dickey_fuller_pvalue}")
        else:
            print(f"Dickey—Fuller test:\tp = {dickey_fuller_pvalue}")

        plt.tight_layout()

    f, ax = plt.subplots()
    f.set_visible(False)
    f.set_figheight(0.3) # figure height in inches
    return
```

```python
for patient_id in res_data['patient_id'].unique():
    patient_df = res_data[res_data['patient_id'] == patient_id]['Glucose'].dropna()
    tsplot(patient_df, lags=4*24, given_id=patient_id)

print('\n\n')
```
